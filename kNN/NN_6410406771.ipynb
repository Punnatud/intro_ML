```python
import numpy as np
import pandas as pd
```

```python
#get all the data from the csv file
df = pd.read_csv('../dataset/breast-cancer-wisconsin.data', header=None)
df.replace('?', 5, inplace=True)
df.drop([0], axis=1, inplace=True)
df
```

```python
#shuffle the data
df = df.sample(frac=1)

#split the data into training and testing
x = np.array(df.drop([10], axis=1),int)
y = np.array(df[10],int)
print(x[0], y[0])
test_size = int(len(x) * 0.3)
print(x.shape, y.shape, test_size)
#split the data into training and testing
x_test = x[:test_size]
y_test = y[:test_size]

x_train = x[test_size:]
y_train = y[test_size:]

print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
```

```python
#finding the distribution of the data
print("from dataset distribution, benign(2): 65.5%, malignant(4): 34.5%")
print("trainning data distribution")
unique, counts = np.unique(y_train, return_counts=True)
print(unique, counts)
print(100*counts[0]/y_train.shape[0], 100*counts[1]/y_train.shape[0])

print("testing data distribution")
unique, counts = np.unique(y_test, return_counts=True)
print(unique, counts)
print(100*counts[0]/y_test.shape[0], 100*counts[1]/y_test.shape[0])
```

```python
def minkowski_distance(x1, x2, p):
    return np.sum(np.abs(x1 - x2) ** p, axis=1) ** (1 / p)

#knn function
def knn(k, p):
    y_pred = []
    for i in range(x_test.shape[0]):
        distance = minkowski_distance(x_train, x_test[i], p)
        distance = np.array(distance)
        distance = np.argsort(distance)
        distance = distance[:k]
        y_pred.append(np.bincount(y_train[distance]).argmax())
    return y_pred
```

```python
#testing the model
y_pred = knn(7, 2)
#compare the prediction with the actual result
countError = 0
for i in range(len(y_pred)):
    if(y_pred[i] != y_test[i]):
        countError += 1
        print("prediction: ",y_pred[i], "actual: ",y_test[i])
print("error: ", countError, "total: ", len(y_pred))
#calculate the accuracy
accuracy = np.sum(y_pred == y_test) / len(y_test)
print("accuracy: ", accuracy)
```
